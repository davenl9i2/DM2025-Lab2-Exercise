## 1. Preprocessing Steps

In the data cleaning phase, I applied minimal preprocessing to allow the BERT model to learn directly from the data context.

1.  **Handling User Mentions (`@user`)**
2.  **Handling URLs**

Through these two processing steps, the model can better learn the distinctive features of each class.

```python
def preprocess_text(text):
    new_text = []
    for t in text.split(" "):

        t = '@user' if t.startswith('@') and len(t) > 1 else t
        
        t = 'http' if t.startswith('http') else t
        new_text.append(t)
    return " ".join(new_text)

train_df['clean_text'] = train_df['text'].apply(preprocess_text)
test_df['clean_text'] = test_df['text'].apply(preprocess_text)
```

## 2. Feature Engineering Steps

In this section, I performed simple label encoding and data splitting. Finally, I utilized the model's inherent tokenization process to generate the BERT Input Features.

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from datasets import Dataset
from transformers import AutoTokenizer

#Label Encoding
le = LabelEncoder()
train_df['label'] = le.fit_transform(train_df['emotion'])

#Save mapping for future inference
id2label = {id: label for id, label in enumerate(le.classes_)}
label2id = {label: id for id, label in enumerate(le.classes_)}
print("Label Mapping:", id2label)

#Stratified Train-Validation Split
train_split_df, valid_split_df = train_test_split(
    train_df, 
    test_size=0.2, 
    stratify=train_df['label'], 
    random_state=42
)

train_dataset = Dataset.from_pandas(train_split_df)
valid_dataset = Dataset.from_pandas(valid_split_df)
test_dataset = Dataset.from_pandas(test_df)

#Tokenization
model_name = "cardiffnlp/twitter-roberta-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)

def preprocess_function(examples):
    return tokenizer(examples["clean_text"], truncation=True, max_length=128)

tokenized_train = train_dataset.map(preprocess_function, batched=True)
tokenized_valid = valid_dataset.map(preprocess_function, batched=True)
tokenized_test = test_dataset.map(preprocess_function, batched=True)
```

## 3. Model Implementation Steps

To tackle the challenges of class imbalance and overfitting, I implemented two custom variations of the RoBERTa trainer.

##### **A. Handling Class Imbalance: Custom Weighted Trainer**
We implemented a `WeightedLossTrainer` to assign higher penalties to minority classes (e.g., 'disgust'). The weights were calculated using **Square Root Smoothing** to ensure the penalties were effective but not overly aggressive.

```python
import torch
from torch import nn
from transformers import Trainer

#Calculate Class Weights
class_counts = train_df['label'].value_counts().sort_index().values
weights = 1.0 / np.sqrt(class_counts)
weights = weights / weights.sum() * len(label2id) # Normalize weights
class_weights = torch.tensor(weights, dtype=torch.float).cuda()

print("Class Weights:", class_weights)

#Custom Trainer with Weighted CrossEntropyLoss
class WeightedTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, **kwargs):
        labels = inputs.get("labels")
        outputs = model(**inputs)
        logits = outputs.get("logits")
        
        loss_fct = nn.CrossEntropyLoss(weight=class_weights)
        loss = loss_fct(logits.view(-1, self.model.config.num_labels), labels.view(-1))
        return (loss, outputs) if return_outputs else loss
```

#### **B. Addressing Overfitting: Fusion Strategy**

To further improve generalization, I developed a `FusionTrainer` that combines **Class Weights** with **Label Smoothing (0.1)**. Additionally, we increased the model's internal **Dropout to 0.2** to prevent memorization of the training data.

```python
from transformers import AutoConfig, AutoModelForSequenceClassification

#Custom Trainer combining Weights and Label Smoothing
class FusionTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, **kwargs):
        labels = inputs.get("labels")
        outputs = model(**inputs)
        logits = outputs.get("logits")
        # Combine Class Weights + Label Smoothing
        loss_fct = nn.CrossEntropyLoss(weight=class_weights, label_smoothing=0.1)
        loss = loss_fct(logits.view(-1, self.model.config.num_labels), labels.view(-1))
        return (loss, outputs) if return_outputs else loss

#Model Initialization
config = AutoConfig.from_pretrained(model_name)
config.hidden_dropout_prob = 0.2        
config.attention_probs_dropout_prob = 0.2 
config.num_labels = 6
config.id2label = id2label
config.label2id = label2id

model = AutoModelForSequenceClassification.from_pretrained(model_name, config=config)
```
